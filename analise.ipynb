```python
from models.jogos import Jogos
import pandas as pd
from pyairtable import Api, Base
import os
from dotenv import load_dotenv

load_dotenv()  # take environment variables from .env.

api_key = os.environ['AIRTABLE_API_KEY']
base_id = 'appt1Ti26Kq8T2LUq'
api = Api(api_key)
base = Base(api, base_id)
table = base.table('Jogos')
table_data = table.all()

lista = []
for r, col in enumerate(table_data): # linha da lista = r , nome da coluna = col
    lista.append(table_data[r]["fields"])

df_jogos = pd.DataFrame(lista)

# df_jogos["Data do Jogo"].dt.year.unique()
df_jogos['Data do Jogo'] = pd.to_datetime(df_jogos['Data do Jogo']) # converte coluna de obj para datetime
cols = df_jogos.columns
cols = cols.map(lambda x: x.replace(' ','_') if isinstance(x, (str)) else x)
df_jogos.columns = cols
```

```python
df_jogos["ano"] = df_jogos["Data_do_Jogo"].dt.year
df_jogos
```

```python
def media_avaliacao_jogos(df, ano=None):
    df = df.loc[df["ano"] == ano] if ano else df
    return df["Avaliação"].mean()
```

```python
media_avaliacao_2025 = media_avaliacao_jogos(df_jogos, 2025)
media_avaliacao_2024 = media_avaliacao_jogos(df_jogos, 2024)
media_avaliacao_geral = media_avaliacao_jogos(df_jogos)
media_avaliacao_2018 = media_avaliacao_jogos(df_jogos, 2018)

mensagem = f"A média de avaliação dos jogos de 2025 é de {media_avaliacao_2025} e de 2024 é de {media_avaliacao_2024}, a avaliação geral é de {media_avaliacao_geral}"

assert mensagem == "A média de avaliação dos jogos de 2025 é de 3.6 e de 2024 é de 3.161764705882353, a avaliação geral é de 3.191780821917808"

print(mensagem)
```
