```python
import pandas as pd

ano = '2017'

onde = 'f'
url = f'https://www.ogol.com.br/equipe/sao-paulo/todos-os-jogos?grp=1&ond={onde}&compet_id_jogos=0&ved=&ano={ano}&comfim=0&equipa_1=2256&menu=allmatches&type=year&op=ver_confronto'
site_resultados = pd.read_html(url,encoding='ISO-8859-1')
resultados_fora = site_resultados[2]

onde = 'c'
url = f'https://www.ogol.com.br/equipe/sao-paulo/todos-os-jogos?grp=1&ond={onde}&compet_id_jogos=0&ved=&ano={ano}&comfim=0&equipa_1=2256&menu=allmatches&type=year&op=ver_confronto'
site_resultados = pd.read_html(url,encoding='ISO-8859-1')
resultados_casa = site_resultados[2]

resultados = pd.concat([resultados_fora, resultados_casa], ignore_index=True)
resultados
```

```python

resultados = resultados.drop(columns=[0,4,8,9])
resultados.columns = ["data","hora", "local", "adversario","resultado_completo","torneio"]
resultados["data_jogo"] = pd.to_datetime(resultados['data'] + ' ' +resultados['hora']).dt.tz_localize('UTC')
```

```python
import numpy as np

# substituir conteúdo nulo da coluna local, para casa (somente neste caso de 2019)
resultados["local"] = resultados["local"].fillna("Casa").replace("(F)","Fora").replace("(C)","Casa")

resultados["gols_pro"] = np.where(resultados["local"] == "Casa", resultados["resultado_completo"].str[0], resultados["resultado_completo"].str[2]) 
resultados["gols_contra"] = np.where(resultados["local"] == "Casa", resultados["resultado_completo"].str[2], resultados["resultado_completo"].str[0]) 
resultados = resultados.drop(columns=["data","hora", "resultado_completo"])

```

```python
tecnicos = pd.read_csv('tecnicos.csv',sep=";")
tecnicos["data_entrada"] = pd.to_datetime(tecnicos["data_entrada"],dayfirst=True).dt.tz_localize('UTC')
tecnicos["data_saida"] = pd.to_datetime(tecnicos["data_saida"],dayfirst=True) + pd.Timedelta(hours=23, minutes=59)
tecnicos["data_saida"] = tecnicos["data_saida"].dt.tz_localize('UTC')
```

```python
# atribui o tecnico conforme a data do jogo e a tabela com as datas de cada técnico
# com o numpy, atualiza o campo técnico conforme data que técnico entrou
# Informação sobre os técnicos https://tricolornaweb.com.br/treinadores-do-spfc-2012-a-2023/ não vi se a API tem dados do técnico

condicoes = [ 
    (resultados["data_jogo"] >= tecnicos.loc[i, 'data_entrada']) & (resultados["data_jogo"] <= tecnicos.loc[i,"data_saida"])
    for i in tecnicos.index
]
opcoes = tecnicos["nome_tecnico"].tolist()

resultados['tecnico'] = np.select(condicoes, opcoes, default=np.nan)

resultados["torneio"] = np.where(resultados["torneio"].str.contains("Brasileirão"), "Serie A", resultados["torneio"]) 
resultados["torneio"] = np.where(resultados["torneio"].str.contains("Copa Sul-Americana"), "CONMEBOL Sudamericana", resultados["torneio"]) 
resultados["torneio"] = np.where(resultados["torneio"].str.contains("Copa do Brasil"), "Copa Do Brasil", resultados["torneio"]) 
resultados["torneio"] = np.where(resultados["torneio"].str.contains("Paulista"), "Paulista - A1", resultados["torneio"]) 


# resultados["torneio"] = "Paulista - A1"
# resultados["torneio"] = "CONMEBOL Sudamericana"
```

```python
import smtplib, ssl
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders

filename = "paulista2018.csv"
# filename = "sulamericana2018.csv"

resultados.to_csv(filename)

# envia o email para importar no airtable

port = 465  # For SSL
sender_email = "gcmoura@gmail.com"
receiver_email = "gcmoura77@hotmail.com"
password = "tjmg bpit igmd kjdu"


message = MIMEMultipart("alternative")
message["Subject"] = "Dados Para importar Airtable"
message["From"] = sender_email
message["To"] = receiver_email

html = f"""\
<html>
  <body>
    <p>Olá,</p>
    <p>
    Segue arquivo .csv para importar no airtable.
    </p>
  </body>
</html>
"""

part1 = MIMEText(html, "html")
message.attach(part1)

# Abrir o ficheiro em modo binário
with open(filename, "rb") as attachment:
    part = MIMEBase("application", "octet-stream")
    part.set_payload(attachment.read())
    
# Codificar ficheiro em caracteres ASCII para enviar emails
encoders.encode_base64(part)

# Adicionar cabeçalho como par de chave/valor ao componente de anexo
part.add_header("Content-Disposition", f"attachment; filename= {filename}")

# Adicionar anexo à mensagem
message.attach(part)
    
# Create a secure SSL context
context = ssl.create_default_context()

with smtplib.SMTP_SSL("smtp.gmail.com", port, context=context) as server:
    server.login(sender_email, password)
    server.sendmail(
        sender_email, receiver_email, message.as_string()
    )
```

## Enviar os arquivos para o Airtable via API

```python
from models.jogos import Jogos
from pyairtable import Api, Base
import os
from dotenv import load_dotenv

load_dotenv()  # take environment variables from .env.


api_key = os.environ['AIRTABLE_API_KEY']
base_id = 'appt1Ti26Kq8T2LUq'
api = Api(api_key)
base = Base(api, base_id)
table = base.table('Jogos')

```

```python
def lista_tabela():
    table_data = table.all()

    lista = []
    for r, col in enumerate(table_data): # linha da lista = r , nome da coluna = col
        lista.append(table_data[r]["fields"])

    df_jogos = pd.DataFrame(lista)

    cols = df_jogos.columns
    cols = cols.map(lambda x: x.replace(' ','_') if isinstance(x, (str)) else x)
    df_jogos.columns = cols

    return df_jogos
```

### Salva registro único

```python
from datetime import datetime

def salvar_registro(adversario, local, gols_pro, gols_contra, torneio, data_jogo, tecnico):
    
    novo_jogo = Jogos(adversario=adversario, 
                        local=local,
                        gols_pro=gols_pro,
                        gols_contra=gols_contra,
                        torneio=torneio,
                        data_jogo=datetime.strptime(data_jogo, "%Y-%m-%d %H:%M:%S"),
                        tecnico=tecnico
                        )

    return novo_jogo.save()

data_str = "2024-11-05 21:30:00"   
if salvar_registro(adversario="Internacional", 
                         local="Casa", 
                         gols_pro=5, 
                         gols_contra=1, 
                         torneio="Serie A", 
                         data_jogo=data_str, 
                         tecnico="Zubeldia"):
    display(lista_tabela())
    
```
