```python
# criação do segundo dataframe
# Adversário, Local, Gols Pró, Gols Contra, Torneio, Resultado, Data do Jogo, Avaliação, Técnico, Pontos

df_final = pd.DataFrame(columns=['adversario', 'local', 'gols_pro', 'gols_contra', 'torneio', 'resultado', 'data_jogo', 'avaliacao', 'tecnico', 'pontos'])

for index, row in df.iterrows():
    
    if row['teams_home_name'] != 'Sao Paulo':
        df_final.loc[index,'adversario'] = row['teams_home_name'] 
        df_final.loc[index,'local'] = 'Fora'
        df_final.loc[index,'gols_pro'] = int(row['goals_away'])
        df_final.loc[index,'gols_contra'] = int(row['goals_home'])

    else:
        df_final.loc[index,'adversario'] = row['teams_away_name'] 
        df_final.loc[index,'local'] = 'Casa'
        df_final.loc[index,'gols_pro'] = int(row['goals_home'])
        df_final.loc[index,'gols_contra'] = int(row['goals_away'])

    if df_final.loc[index,'gols_pro'] > df_final.loc[index,'gols_contra']:
        df_final.loc[index,'resultado'] = 'Vitória'
        df_final.loc[index,'pontos'] = 3
    elif df_final.loc[index,'gols_pro'] < df_final.loc[index,'gols_contra']:
        df_final.loc[index,'resultado'] = 'Derrota'
        df_final.loc[index,'pontos'] = 0
    else:
        df_final.loc[index,'resultado'] = 'Empate'
        df_final.loc[index,'pontos'] = 1
    
    df_final.loc[index,'torneio'] = row['league_name']
    # df_final.loc[index,'data_jogo'] = datetime.datetime.fromisoformat(row['fixture_date'])
    df_final.loc[index,'data_jogo'] = row['fixture_date']

df_final = df_final.astype({"pontos": int, "gols_contra": int, "gols_pro": int}) # converte coluna pontos de obj para int
df_final['data_jogo'] = pd.to_datetime(df_final['data_jogo']) # converte coluna de obj para datetime
df_final.head()
```

```python
# com o numpy, atualiza o campo técnico conforme data que Dorival Junior entrou
# Informação sobre os técnicos https://tricolornaweb.com.br/treinadores-do-spfc-2012-a-2023/ não vi se a API tem dados do técnico
df_final["tecnico"] = np.where(df_final["data_jogo"] < '2023-04-22','Rogério Ceni','Dorival Junior')
display(df_final)
```
